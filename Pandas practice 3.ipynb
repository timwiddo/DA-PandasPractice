```python
pd.crosstab(df2.Nationality, df2.Handedness)
```

```python
pd.crosstab(df2.Sex, df2.Handedness, margins = True)
```

```python
pd.crosstab(df2.Sex, [df2.Handedness, df2.Nationality], margins = True)
```

```python
import numpy as np
# avg age of f/m who are left/right handed through vals = age and aggfunc = avg
pd.crosstab(df2.Sex, df2.Handedness, values = df2.Age, aggfunc = np.average)
```
