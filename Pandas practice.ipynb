## Dataframes creation

````python
# Dictionaries can become a df too

weather_data = {
    'day': ['1/1/2024', '1/2/2024', '1/3/2024'],
    'temperature': [32, 35, 28],
    'windspeed': [6, 7, 2],
    'event': ['Rain', 'Sunny', 'Snow']   
    }

df = pd.DataFrame(weather_data)
df
````

## Tuple Lists

````python
weather_data_list = [
    ('1/1/2024',32, 6, 'Rain'),
    ('1/2/2024',35, 7, 'Sunny'),
    ('1/3/2024',28, 2, 'Snow')
    ]

df = pd.DataFrame(weather_data_list, columns = ['day', 'temperature', 'windspeed', 'event'])
df
````

## Dictionary List

````python
weather_dict_list = [
    {'day': '1/1/2024', 'temperature': 32, 'windspeed': 6, 'event': 'Rain' },
    {'day': '1/2/2024', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny' },
    {'day': '1/3/2024', 'temperature': 28, 'windspeed': 2, 'event': 'Snow' }
]

df = pd.DataFrame(weather_dict_list)
df
````

````python
# inplace true makes changes inside of original df
# missing_values can also be a dictionary
import numpy as np

missing_values = {
    'eps': ['n.a.', 'not available'],
    'revenue': ['n.a.', 'not available', -1],
    'price': ['n.a.', 'not available'],
    'people': ['n.a.', 'not available']
     }

df.replace(missing_values, np.nan, inplace=True)
df
````

## Replacement Mapping

````python
df3 = pd.DataFrame({
    'score': ['exceptional', 'average', 'good', 'poor', 'average', 'exceptional'],
    'Name': ['Tim', 'Dew', 'Steve', 'Markus', 'Niklas', 'Soheil']
})
df3
````

````python
df3_rep = df3.replace(['poor', 'average', 'good', 'exceptional'], [4, 3, 2, 1])
df3_rep
````

````python
df3_rep.rename(columns = {'score': 'grade'}, inplace = True)
df3_rep
````
